# Context
## What is the object and its history?
My object is the 'Journal of a Continental Tour' created between the years of 1867-1868. This 19th century travel journal consists of journal entries, sketches, photographs, and coloured sketches. 

This object can be found in the University of Glasgow's special collections and archives under the reference: GB 247 MS Gen 13. The journal was acquired for the university by John Smith on the date 18th November 1953.

It can also be found at this link: https://www.gla.ac.uk/collections/#/details?irn=250262&catType=C&referrer=/results&q=ms+gen+13

## How will a new perspective be introduced and how will this project engage with audiences?
This project will engage with audiences by introducing more enhanced images from the journal. Usually at a cultural heritage site the general public would not sit and read through an entire journal, by enhancing some of the images whilst including details from the written text, users can experience the journal within a much smaller timeframe. This will make sure that a larger audience can see the journal without having to leaf through it, but have a higher chance of enjoying the product. Also, with an object this old and fragile having a large quantity of people leafing through the pages can create more damage to the physical object. With a project like this physical contact is reduced and also in a way immortalises the journal, so people can always have an idea of what the journal is like in a more engaging way. This way of seeing the journal may also be more appealing to younger audiences by looking at photos and hearing audio may engage them more than a large volume of text.

My objectives for this project are to:

- Also using the journal to add details to the images using AI.

# Design Approach
## How this design incorporates generative AI models as part of the human creative process.
This project uses generative AI models such as Paintify, Photoshop, and ChatGPT to create pieces of artwork and text that enhance the human experience by engaging them with cultural heritage, more sepcifically a 19th century travel journal.These resources will be used alongside human creativity to produce enhanced images from this journal. The journal currently comprises of large amounts of text with few sketches and drawings involved. This project will use existing sketches and use generative AI to produce more detailed cohesive versions of these. Once transformed using generative AI tools, human creativity will be used alongside tools in Photoshop to enhance them even further by adding details from the sketches and text that the AI tools may have missed or skipped over.

## Tools Used

ChatGPT used for the new journal entries objective. Paintify, Photoshop, and DALL-E used for enhancing and editing images. PlayHT used for generating audio files.

### How were these tools used?
#### Canva, Paintify, and DALL-E
I used Canva by using the application Paintify. The application Paintify uses an uploaded image and transforms it - using generative AI - into a painting. I used this resource for the image of the horse-drawn coach. I uploaded the initial sketch and came out with an image that enhanced the sketch into a significantly more detailed image with colour and finished drawings.
Paintify will work by taking my input - the drawn sketch - and applying their algorithm to it which will produce the painting-like image. They will have developed this tool by inputting lots of training data to train the AI into doing this, to then output a painting image from the input image.
I also used the application DALL-E through Canva to create my own prompts to then generate images.

#### Photoshop
I used photoshop to then add further detail to images produced from Paintify. For example, the image of the coach through paintify did not include any luggage on top of the coach. I then used photoshop and used the generative fill function to get luggage added to the top. I also used photoshop to create wider scenes in images, taking one of the drawings and giving it a background within the context of the journal entries.
Photoshop will use generative AI to fill in the places I want to add detail. By selecting an area and then adding a prompt the generative fill takes your input and fits in your prompt to the area you have selected. This tool will have been trained using training data to take inputs and create whatever it is that is asked.

#### ChatGPT
I used ChatGPT to generate new journal entries from the original entries. The application took the journal entries I pasted into the input box and generated a new journal entry in the style of the original journal. Using generative AI the platform is able to take the original text and how it is written and produce new text. It will have learnt how to do this by using training data.

#### PlayHT
I used PlayHT to generate audio, this was done by entering an excerpt of the journal entry and choosing the type of voice I wanted, and then an audio was generated.

## Method
1. First I came up with my project proposal - I decided that I would attempt to create animations from the sketches and enhanced images. From this project proposal I received feedback, such as would I consider using audio with the animations.
2. From here I decided I would only use generative AI to create enhanced images, and some samples of audio to go along with some of the images. 
3. Then I trialed and experimented with different methods of creating AI generated images using Photoshop and Canva. Then I trialed with PlayHT to create generated audio from excerpts of the journal.
4. I then went through the transcript of the travel journal to gather context clues for the sketches. I selected drawings and got an idea of the image I wanted to create using generative AI.
5. Initially, I chose the sketch of the woman and man packing their luggage; the sketch of the coach with the luggage on top; the moustached man; and the coloured drawing of the girl.
6. From here I began with the coloured drawing of the girl. This drawing of the girl was the daughter of a woman named Mrs Keith, who the author saw when picking up Margaret from a children's party. Using this context, I used Photoshop to generate a background for the young girl. I used the prompt "party scene in 19th century" first, and then used "child's party scene in 19th century". The first prompt came up with promising results and generated other guests at the party. The second prompt however, resulted in pictures with guests not having human heads. The first prompt was not perfect as one of the people in my chosen photo has three hands. However, I think this enhanced image is useful in providing context for the original drawing and how the real life scene might have looked.
7. Next, I began with the drawing of the horse-drawn coach with luggage on top. First, I attempted to use Photoshop to transform this sketch into a detailed drawing but failed, the generative fill did not seem to work when trying to prompt it into making sections of the sketch into a real image. So, next I used the Paintify app on Canva to upload the image of the sketch and it created a good idea of what the coach might have looked like in real life, albeit in painting form. A detail missing that I found important to the original sketch was the luggage on top. So, I then took the painting to Photoshop where I prompted the generative fill to create luggage on top of the coach. Using a combination of the 2 tools worked perfectly in creating and setting a scene of what the author of the travel journal might have seen in real life.
8. Next, I took the drawing of the moustached man and used Paintify to create a painting version of the man, this lacked a background. So, instead I used a different method and used DALL-E to generate a picture using the description: "a man with a large sandy grey moustache and a top hat, also in a suit stood in a train station with some women in the 19th century". This descriptions created an image with resembled the drawing but also provided a background and used the context from the journal entries to create a larger image.
9. I next used DALL-E to create an image of the couple packing their suitcases. I used the prompt "a couple in distress packing 3 suitcases with books and clothes strewn across the floor in 1867". This produced two different images that were both suitable and resembled the drawing very well.
10. I used PlayHT to input the paragraph of text from the transcript about the Girl at the party and used it to generate an audio to go along with the generated image.
